<h2>Parte de orientação a objetos:</h2>
<p>A ideia é criar um DataFrame a partir de um dicionário em que suas chaves são os atributos do objeto 'Vendas' e os seus valores estão em um formato de Dicionários separados.<br><br>Coloquei os atributos protegidos e fazendo getters e setters.</p>


```python
class Vendas:
    def __init__(self,marca='',ano='',km='',cor='',motor='',combustivel='',cambio=''):
        
        ## ATRIBUTOS PROTEGIDOS
        
        self._marca = []
        self._ano = []
        self._km = []
        self._cor = []
        self._motor = []
        self._combustivel = []
        self._cambio = []
        
        
        ## GETTERS E SETTERS DE CADA ATRIBUTO
    def get_marca(self):
        return self._marca
    def set_marca(self,marca):
        self._marca.append(marca)
    
    def get_ano(self):
        return self._ano
    def set_ano(self,ano):
        self._ano.append(ano)
        
    def get_km(self):
        return self._km
    def set_km(self,km):
        self._km.append(km)
        
    def get_cor(self):
        return self._cor
    def set_cor(self,cor):
        self._cor.append(cor)
        
    def get_motor(self):
        return self._motor
    def set_motor(self,motor):
        self._motor.append(motor)
        
    def get_combustivel(self):
        return self._combustivel
    def set_combustivel(self,combustivel):
        self._combustivel.append(combustivel)     
        
    def get_cambio(self):
        return self._cambio
    
    def set_cambio(self,cambio):
        self._cambio.append(cambio)
                
   
        
        
```

<h2>Os arquivos são dividos em dois tipos de formato.</h2>
<p>Pra eu reconhecer qual formato eu decidi reconhecer pela primeira linha, dando um split na linha para saber quantas palavras tem na linha, podendo ter tamanho exatamente 1 (caso do test6 ao test12) e tendo mais de 1 que é o resto.
    <br><br> Irei chamar de 'caso_1' o caso especial de test6 até test12, e de 'caso_2' os demais casos. </p>

```python
v = Vendas()

for file in file_lines:
    stop = False
    caso_1 = False
    caso_2 = False
    for i in range(len(file_lines[file])): ## Cada iteração desse for é uma linha
        linha = file_lines[file][i]
        linha_split = linha.split()
        size = len(linha_split) ## Saber tamanho da lista
        if i==0:
            if size == 1:
                caso_1 = True    #test6 <--->test12
            else:
                caso_2 = True    #resto           
        if linha.title() == 'Ano':
            ano = file_lines[file][i+1].split('/')
            x = ano[0]
            x = x.replace(' ','')
            v.set_ano(x)
            
        elif linha.title() == 'Km' or linha.title()== 'Quilometragem':
            km = int(file_lines[file][i+1])
            v.set_km(km)
            
        elif linha.title() == 'Cor':
            cor = file_lines[file][i+1]
            v.set_cor(cor)     
                
        if caso_2: 
            if i==0:  ## setar Marca no caso 2 
                marca = linha_split[0]
                v.set_marca(marca)
            elif linha_split[0].title() == 'Conservação:':
                for y in range(size):
                    if y != 0:
                        if linha_split[y-1] == 'Combustível:':
                            combustivel = linha_split[y]
                            v.set_combustivel(combustivel)
                        elif linha_split[y-1] == 'Motor:':
                            motor = linha_split[y]
                            motor = float(motor)
                            v.set_motor(motor)
                        elif linha_split[y-1] == 'Transmissão:':
                            cambio = linha_split[y]
                            v.set_cambio(cambio)
                                                         
        else:
            if i != 0:
                if file_lines[file][i-1] == 'Combustível':
                    combustivel = linha
                    v.set_combustivel(combustivel)
                elif file_lines[file][i-1] == 'Potência do motor':
                    motor = linha
                    motor = float(motor)
                    v.set_motor(motor)
                elif file_lines[file][i-1] == 'Câmbio':
                    cambio = linha
                    v.set_cambio(cambio)
                elif file_lines[file][i-1] == 'Marca':
                    marca = linha
                    v.set_marca(marca)                    
```

<h2>Dando replace na marca com dois nomes:</h2>

```python
df['_marca'] = df['_marca'].replace('GM - CHEVROLET','Chevrolet')
df
```

```python
df_35kkm = df[df['_km']<35000]
values = df_35kkm['_km'].values
index = df_35kkm['_marca'].values
plt.style.use('dark_background')
plt.ylabel('Quilometragem',fontdict={'color':'white','fontsize':18})
plt.xlabel('Marca',fontdict={'color':'white','fontsize':18})
plt.bar(index,values)
plt.title('Carros abaixo de 35 mil quilômetros',fontdict={'color':'white','fontsize':20})
plt.savefig('carros35mil.png')
```

```python
labels = df['_cor'].value_counts().index
values = df['_cor'].value_counts()
plt.style.use('ggplot')
plt.pie(values,labels = labels,explode = (0.05, 0.05, 0.05, 0.05),autopct = '%1.1f%%',shadow = True,startangle = 0,wedgeprops={'alpha':0.5},textprops={"fontsize":20,"color":'green'},colors=['white','black','silver','blue'])
plt.title('Porcentagem das cores dos carros',fontdict={'color':'black','fontsize':20})
plt.savefig('corescarros.png',facecolor='white')
```

```python
y = df['_km'].sort_values().values
x = np.arange(15)
plt.tick_params(labelbottom=False,width=5,pad=0.2,zorder=4)
plt.style.use('ggplot')
plt.ylabel('Quilômetros rodados',fontdict={'color':'black','fontsize':18})
plt.scatter(x,y)
plt.title('Quilometragem dos carros',fontdict={'color':'black','fontsize':20})
plt.savefig('quilometragem_carros.png',facecolor='white')
```

```python
ano = df['_ano'].value_counts()
ano = ano.sort_index(ascending=False)
x = ano.index
y = ano.values
plt.style.use('dark_background')
plt.barh(x,y)
plt.xticks(np.arange(0, 5, step=1))
plt.xlabel('Quantidade')
plt.ylabel('Ano')
plt.title('Quantidade de carros de cada ano')
plt.savefig('qtd_ano.png')
```

```python
motor = df['_motor'].value_counts()
motor = motor.sort_index()
values = motor.values
labels = motor.index
plt.style.use('ggplot')
plt.pie(values,autopct = '%1.1f%%',shadow = True,startangle = 0,wedgeprops={"ec":"k"},textprops={"fontsize":20,"color":'black'})
plt.legend(labels = labels,edgecolor='yellow',facecolor ='black')
plt.title('Porcentagem dos tipos de motores',fontdict={'color':'black','fontsize':20})
plt.savefig('tipo_motores.png',facecolor='white')
```
